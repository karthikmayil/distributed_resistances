```python
import pandas as pd
import numpy as np
import os
import proplot as pplt
```

```python
fig,axs = pplt.subplots(ncols=4,nrows=1,sharex=False,sharey=False,facecolor='w')

NJ_sep = 22
NJ = 33
NJ_cathode = NJ - NJ_sep

for i in range(4):
    filepath_TVT = f'0.5C/{i}/Time_Voltage.txt'
    TVT = pd.read_csv(filepath_TVT,delim_whitespace=True,skiprows=[1])
    
    if len(TVT)<20:
        continue
    
    filepath_TVPT = f'0.5C/{i}/Time_Voltage_Position.txt'
    TVPT = pd.read_csv(filepath_TVPT,delim_whitespace=True,skiprows=[1])
    t = TVPT.Time.unique()[-3]
    tdf = TVPT.loc[TVPT.Time==t]
    
    ax=axs[0]
    ax.plot(tdf.Position.values[-NJ_cathode-1:],tdf.Solid_Conc.values[-NJ_cathode-1:],'o')
    
    ax=axs[1]
    ax.plot(TVT.Time.values,TVT.Voltage.values,'-')
    
    ax=axs[2]
    ax.plot(tdf.Position.values[-NJ_cathode-1:],tdf.Solution_Pot.values[-NJ_cathode-1:],'o')
    
    ax=axs[3]
    ax.plot(tdf.Position.values[-NJ_cathode-1:],tdf.Voltage.values[-NJ_cathode-1:],'o')

axs[1].format(ylabel='Voltage',xlabel='Time')
axs[0].format(ylabel='x in LixNMC',xlabel='Position')    
```
